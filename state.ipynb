```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 统计每个桶的命中率
def get_vec(path, mode=float):
  items = None
  fb = open(path, "r")
  lines = fb.readlines()
  for line in lines:
    if line is None:
      break
    line_split = line.split(' ')[:-1]
    items = [mode(i) for i in line_split]
  fb.close()
  return items
```

```python
x = get_vec('/tmp/radius_uc.txt')
y = get_vec('/tmp/radius_rc.txt')
```

```python
a = plt.hist(x, bins=200)
# y = plt.hist(y, bins=200)
```
